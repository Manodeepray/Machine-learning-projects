Linear Regression:

Characteristics: Linear relationship between dependent and independent variables.
Variables: One dependent variable, one or more independent variables.
Real-Life Applications: Predicting house prices based on square footage, forecasting sales based on advertising spending.

Multiple Regression:

Characteristics: Extension of linear regression for multiple independent variables.
Variables: One dependent variable, two or more independent variables.
Real-Life Applications: Predicting a student's GPA based on study hours, class attendance, and previous grades.

Polynomial Regression:

Characteristics: Fits a polynomial equation to the data.
Variables: One dependent variable, one independent variable (or more with higher degrees).
Real-Life Applications: Modeling the relationship between temperature and gas consumption.

Support Vector Machine (SVM) Regression:

Characteristics: Finds a hyperplane that best represents the data.
Variables: One dependent variable, one or more independent variables.
Real-Life Applications: Stock price prediction, predicting patient outcomes in healthcare.

Decision Tree Regression:

Characteristics: Uses a tree-like model to make decisions.
Variables: One dependent variable, one or more independent variables.
Real-Life Applications: Predicting customer churn, determining salary based on education and experience.

Random Forest Regression:

Characteristics: Ensemble of decision trees for improved accuracy.
Variables: One dependent variable, one or more independent variables.
Real-Life Applications: Predicting customer preferences in marketing, estimating crop yields in agriculture.

**evaluating regression model**

Evaluation metrics;

* Mean/Median of prediction
* Standard Deviation of prediction
* Range of prediction
* Coefficient of Determination (R2)
* Relative Standard Deviation/Coefficient of Variation (RSD)
* Relative Squared Error (RSE)
* Mean Absolute Error (MAE)
* Relative Absolute Error (RAE)
* Mean Squared Error (MSE)
* Root Mean Squared Error on Prediction (RMSE/RMSEP)
* Normalized Root Mean Squared Error (Norm RMSEP)
* Relative Root Mean Squared Error (RRMSEP)


**multiple linear regression**

```python
dataset1 = pd.read_csv('/content/drive/MyDrive/modeol selection/Machine Learning A-Z (Model Selection)/Regression/Data.csv')
X1 = dataset1.iloc[:, :-1].values
y1 = dataset1.iloc[:, -1].values

from sklearn.model_selection import train_test_split
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size = 0.2, random_state = 0)

from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X1_train, y1_train)

y1_pred = regressor.predict(X1_test)
np.set_printoptions(precision=2)
print(np.concatenate((y1_pred.reshape(len(y1_pred),1), y1_test.reshape(len(y1_test),1)),1))

from sklearn.metrics import r2_score
mlr=r2_score(y1_test, y1_pred)
mlr
```

**polynomial regression**

```python
dataset2 = pd.read_csv('/content/drive/MyDrive/modeol selection/Machine Learning A-Z (Model Selection)/Regression/Data.csv')
X2 = dataset2.iloc[:, :-1].values
y2 = dataset2.iloc[:, -1].values

from sklearn.model_selection import train_test_split
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size = 0.2, random_state = 0)

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
poly_reg = PolynomialFeatures(degree = 4)
X2_poly = poly_reg.fit_transform(X2_train)
regressor = LinearRegression()
regressor.fit(X2_poly, y2_train)

y2_pred = regressor.predict(poly_reg.transform(X2_test))
np.set_printoptions(precision=2)
print(np.concatenate((y2_pred.reshape(len(y2_pred),1), y2_test.reshape(len(y2_test),1)),1))

from sklearn.metrics import r2_score
plr=r2_score(y2_test, y2_pred)
plr
```

**svm regression**

```python
dataset3 = pd.read_csv('/content/drive/MyDrive/modeol selection/Machine Learning A-Z (Model Selection)/Regression/Data.csv')
X3 = dataset3.iloc[:, :-1].values
y3 = dataset3.iloc[:, -1].values

y3 = y3.reshape(len(y3),1)

from sklearn.model_selection import train_test_split
X3_train, X3_test, y3_train, y3_test = train_test_split(X3, y3, test_size = 0.2, random_state = 0)

from sklearn.preprocessing import StandardScaler
sc_X3 = StandardScaler()
sc_y3 = StandardScaler()
X3_train = sc_X3.fit_transform(X3_train)
y3_train = sc_y3.fit_transform(y3_train)

from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X3_train, y3_train)

y3_pred = sc_y3.inverse_transform(regressor.predict(sc_X3.transform(X3_test)).reshape(-1,1))
np.set_printoptions(precision=2)
print(np.concatenate((y3_pred.reshape(len(y3_pred),1), y3_test.reshape(len(y3_test),1)),1))

from sklearn.metrics import r2_score
svmr=r2_score(y3_test, y3_pred)
svmr
```

**decision tree regression**

```python
dataset4 = pd.read_csv('/content/drive/MyDrive/modeol selection/Machine Learning A-Z (Model Selection)/Regression/Data.csv')
X4 = dataset4.iloc[:, :-1].values
y4 = dataset4.iloc[:, -1].values

from sklearn.model_selection import train_test_split
X4_train, X4_test, y4_train, y4_test = train_test_split(X4, y4, test_size = 0.2, random_state = 0)

from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X4_train, y4_train)

y4_pred = regressor.predict(X4_test)
np.set_printoptions(precision=2)
print(np.concatenate((y4_pred.reshape(len(y4_pred),1), y4_test.reshape(len(y4_test),1)),1))

from sklearn.metrics import r2_score
dtr=r2_score(y4_test, y4_pred)
dtr
```

**random forest regression**

```python
dataset5 = pd.read_csv('/content/drive/MyDrive/modeol selection/Machine Learning A-Z (Model Selection)/Regression/Data.csv')
X5 = dataset5.iloc[:, :-1].values
y5 = dataset5.iloc[:, -1].values

from sklearn.model_selection import train_test_split
X5_train, X5_test, y5_train, y5_test = train_test_split(X5, y5, test_size = 0.2, random_state = 0)

from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 10, random_state = 0)
regressor.fit(X5_train, y5_train)

y5_pred = regressor.predict(X5_test)
np.set_printoptions(precision=2)
print(np.concatenate((y5_pred.reshape(len(y5_pred),1), y5_test.reshape(len(y5_test),1)),1))

from sklearn.metrics import r2_score
rfr=r2_score(y5_test, y5_pred)
rfr
```

```python
print("r2 score of the following:")
print("multiple linear regression",mlr)
print("polynomial linear regression",plr)
print("svm regression",svmr)
print("decision tree regression",dtr)
print("random forest regression",rfr)
```
