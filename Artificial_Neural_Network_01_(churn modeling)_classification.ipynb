```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
tf.__version__
```

```python
dataset = pd.read_csv('/content/drive/MyDrive/machine learning udemy r/Machine Learning A-Z (Codes and Datasets)/Part 8 - Deep Learning/Section 39 - Artificial Neural Networks (ANN)/Python/Churn_Modelling.csv')
```

```python
x = dataset.iloc[:,3:-1].values
y = dataset.iloc[:,-1].values

```

```python
#label encoding the gender column ie 0 and 1

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[:,2] = le.fit_transform(x[:,2])

x
```

```python
#now one hot encodeing the geography

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')# 1 is the column index
x = np.array(ct.fit_transform(x))
x
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
X_train
```

```python
#FEATURE SCALING : IT IS ABSOLUTELY NECESSARY TO FEATURE SCALE FOR DEEP LEARNING RERGARDLESS OF 0 AND 1
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

```

```python
X_train
```

```python
X_test
```

#BUILDING THE ANN

```python
#CREATE THE ANN

#initialize

ann = tf.keras.models.Sequential()
```

```python
#add input layer and first hidden layer

ann.add(  tf.keras.layers.Dense( units = 6 , activation = 'relu')) #adding the dense class

#units is the no of neurons in ffirst hidden layer...random no rule
#activation = '' == activation function layer
#relu = rectifier activation function

#add 2nd hidden layer

ann.add(  tf.keras.layers.Dense( units = 6 , activation = 'relu'))

#change the no . of units for diff result

```

```python
#output layer

ann.add(  tf.keras.layers.Dense( units = 1 , activation = 'sigmoid'))


#since we need only binary out put like 1 or 0 we need only one neuron ie dimension
#but if it was classification with 2 or more classes like a b and c, we would need 3 o/p neuron ie 3 dimensions
#and we would need to one hot encode the three classes as well like a= 100 , b =010 and c = 001,,so on
```

#TRAINING THE ANN

```python
#compile the ann




ann.compile( optimizer = 'adam' , loss = 'binary_crossentropy' , metrics = ['accuracy'])

#optimizer = 'adam' best for schtocastic gdm
#loss = 'binary_crossentropy' for binary classification
#loss = 'categorical_crossentropy' for non-binary classification

#and for non-binary classification with more than 2 classification the o/p activation functoin should be 'softmax' not 'sigmoid'
#metrics = ['accuracy']  for evaluation





```

```python
#Training the ANN on the Training set





ann.fit(X_train, y_train, batch_size = 32, epochs = 100)
```


Use our ANN model to predict if the customer with the following informations will leave the bank:

```python
#prediction > 0.5 == 1 bcoz of sigmoid function

#here false means the customer will not leave the bank
```

Important note 1: Notice that the values of the features were all input in a double pair of square brackets. That's because the "predict" method always expects a 2D array as the format of its inputs. And putting our values into a double pair of square brackets makes the input exactly a 2D array.

Important note 2: Notice also that the "France" country was not input as a string in the last column but as "1, 0, 0" in the first three columns. That's because of course the predict method expects the one-hot-encoded values of the state, and as we see in the first row of the matrix of features X, "France" was encoded as "1, 0, 0". And be careful to include these values in the first three columns, because the dummy variables are always created in the first columns.

```python
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
y_pred
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))

```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
