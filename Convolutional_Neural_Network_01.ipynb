preproessing the training set


```python
#preprocessing the data

train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)
#overfitting ^


training_set = train_datagen.flow_from_directory(#apply the train datagen
        '/content/drive/MyDrive/machine learning udemy r/Machine Learning A-Z (Codes and Datasets)/Part 8 - Deep Learning/Section 40 - Convolutional Neural Networks (CNN)/Python/CNN datase(cats and dogs)/dataset/training_set',
        target_size=(64, 64),#final size of pic that would be fed to cnn
        batch_size=32,
        class_mode='binary')#orcategorical (more than 2 classes)


```

preprocessing the test set

```python
test_datagen = ImageDataGenerator(rescale=1./255)#for test set needs freature scaling

test_set = test_datagen.flow_from_directory(
        '/content/drive/MyDrive/machine learning udemy r/Machine Learning A-Z (Codes and Datasets)/Part 8 - Deep Learning/Section 40 - Convolutional Neural Networks (CNN)/Python/CNN datase(cats and dogs)/dataset/test_set',
        target_size=(64, 64),
        batch_size=32,
        class_mode='binary')

```

**CNN**

```python
#initialize the cnn
cnn = tf.keras.models.Sequential()
```

first layer


```python
#convolution

cnn.add(tf.keras.layers.Conv2D( filters = 32 , kernel_size = 3 , activation = 'relu' , input_shape = [64 ,64 ,3] )) #kernel_size = row of feature dectector(square array) filters = no of nodes



#pooling(MAX POOLING)


cnn.add(tf.keras.layers.MaxPool2D( pool_size = 2 , strides = 2))#pool size the size of frame , stride = number of pixels frame shifted

```

adding 2nd layer


```python
cnn.add(tf.keras.layers.Conv2D( filters = 32 , kernel_size = 3 , activation = 'relu' ,))# 'input_shape = [64 ,64 ,3]' only needed for first layer

cnn.add(tf.keras.layers.MaxPool2D( pool_size = 2 , strides = 2))
```

flattening

```python
cnn.add(tf.keras.layers.Flatten())
```

full connection

```python
#connecting the 1d flattened layer as input

cnn.add(  tf.keras.layers.Dense( units = 128 , activation = 'relu'))
```

output layer

```python
cnn.add(  tf.keras.layers.Dense( units = 1 , activation = 'sigmoid'))

```

**training the CNN**

```python
#compile
cnn.compile( optimizer = 'adam' , loss = 'binary_crossentropy' , metrics = ['accuracy'])

#trining the cnn on training set and evaluating it on the test set


cnn.fit( x = training_set , validation_data = test_set , epochs = 25)

```
