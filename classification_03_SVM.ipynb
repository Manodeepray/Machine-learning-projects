```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

dataset = pd.read_csv('/content/drive/MyDrive/machine learning udemy r/Machine Learning A-Z (Codes and Datasets)/Part 3 - Classification/Section 16 - Support Vector Machine (SVM)/Python/Social_Network_Ads.csv')
```

```python
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split( X, y ,test_size =0.25, random_state = 0 )
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.svm import SVC
classifier = SVC( kernel = 'linear' , random_state = 0)
```

```python
classifier.fit(X_train , y_train)
```

```python
y_predict = classifier.predict(X_test)
```

```python
y_predict
```

```python
## to test a value age 30 income 87000
print(classifier.predict(sc.transform([[30,87000]])))
```

```python
#concatenation of the ypred and y test

print( np.concatenate ( ( y_predict.reshape(len(y_predict),1) , y_test.reshape(len(y_test),1) ),1  )  )
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_predict)
print(cm)
accuracy_score(y_test, y_predict)
```
